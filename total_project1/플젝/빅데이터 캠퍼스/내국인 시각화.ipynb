```python
import pandas as pd
```

```python
hap18 = pd.read_csv("E:/oyj/result2/18hap_topupjong_local.csv")
hap18['year'] = 2018
```

```python
hap19 = pd.read_csv("E:/oyj/result2/19hap_topupjong_local.csv")
hap19['year'] = 2019
hap20 = pd.read_csv("E:/oyj/result2/20hap_topupjong_local.csv")
hap20['year'] = 2020
```

```python
hap = pd.concat([hap18,hap19,hap20])
hap['카드이용금액계'] = hap['카드이용금액계']/10000000000
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib
import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
matplotlib.rc('font', family = font_name)
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
import numpy as np

label = hap['year'].unique()
index = np.arange(len(label))
legend = hap['내국인업종코드'].unique()
alpha = 0.4
```

```python
new_legend = hap['내국인업종코드'][:8]
```

```python
grape_01 = list(hap[hap['내국인업종코드'] == new_legend[0]]['카드이용금액계'])
grape_02 = list(hap[hap['내국인업종코드'] == new_legend[1]]['카드이용금액계'])
grape_03 = list(hap[hap['내국인업종코드'] == new_legend[2]]['카드이용금액계'])
grape_04 = list(hap[hap['내국인업종코드'] == new_legend[3]]['카드이용금액계'])
grape_05 = list(hap[hap['내국인업종코드'] == new_legend[4]]['카드이용금액계'])
grape_06 = list(hap[hap['내국인업종코드'] == new_legend[5]]['카드이용금액계'])
grape_07 = list(hap[hap['내국인업종코드'] == new_legend[6]]['카드이용금액계'])
grape_08 = list(hap[hap['내국인업종코드'] == new_legend[7]]['카드이용금액계'])

bottom1 = grape_01
bottom2 = [x+y for x,y in zip(bottom1,grape_02)]
bottom3 = [x+y for x,y in zip(bottom2,grape_03)]
bottom4 = [x+y for x,y in zip(bottom3,grape_04)]
bottom5 = [x+y for x,y in zip(bottom4,grape_05)]
bottom6 = [x+y for x,y in zip(bottom5,grape_06)]
bottom7 = [x+y for x,y in zip(bottom6,grape_07)]
bar = [int(x+y) for x,y in zip(bottom7,grape_08)]

alpha = 0.7
p01 = plt.bar(label,grape_01,alpha = alpha)
p02 = plt.bar(label,grape_02, bottom =bottom1,alpha = alpha)
p03 = plt.bar(label,grape_03, bottom =bottom2,alpha = alpha)
p04 = plt.bar(label,grape_04, bottom =bottom3,alpha = alpha)
p05 = plt.bar(label,grape_05, bottom =bottom4,alpha = alpha)
p06 = plt.bar(label,grape_06, bottom =bottom5,alpha = alpha)
p07 = plt.bar(label,grape_07, bottom =bottom6,alpha = alpha)
p08 = plt.bar(label,grape_08, bottom =bottom7,alpha = alpha)
plt.plot(label,bar, color = 'black')

plt.title('업종별(8개) 내국인 사용내역',fontsize =20)
plt.ylabel('카드이용금액(백억)',fontsize =15)
plt.xlabel('Year',fontsize =15)
plt.legend((p01[0],p02[0],p03[0],p04[0],p05[0],p06[0],p07[0],p08[0]),new_legend, bbox_to_anchor=(1,0.5))
plt.xticks(label)
plt.yticks([0,1000,2000,3000,4000,5000,6000,7000])

for i, v in enumerate(label):
    plt.text(v,bar[i],bar[i],
            fontsize=15,
            color='black',
            horizontalalignment='center',
            verticalalignment='bottom')

plt.figure(figsize=(25,22))
plt.show()
```
